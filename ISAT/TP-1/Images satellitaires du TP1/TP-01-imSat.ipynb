3-1) Cette partie de code sest surtout bas sur lutilisation des compositions colores Rouge (R), Vert

##### A-composition TM3/TM2/TM1 "COMPOSITION COLORE VRAIES COULEURS"
Composition colore 321 (couleurs naturelles) 
Pour pouvoir procder au contrle de la ralit du terrain, la composition 321 qui 
correspond aux couleurs naturelles, a t utilise comme le montre limage resultante suivante



```python
sizex=400 #largeur
sizey=800 #longeur

imgRGB=np.zeros([sizex, sizey,3]) #creation d'une matrice d'une image initial


IMG3=plt.imread('TM3.BMP') #lecture de TM3.BMP au variable IMG3
IMG2=plt.imread('TM2.BMP') #lecture de TM2.BMP au variable IMG2
IMG1=plt.imread('TM1.BMP') #lecture de TM1.BMP au variable IMG1


#assignation des image au different niveau de couleur RGB du imgRGB
imgRGB[:,:,0]=IMG3[:,:,0]/255 #niveau rouge
imgRGB[:,:,1]=IMG2[:,:,1]/255 #niveau vert
imgRGB[:,:,2]=IMG1[:,:,2]/255 #niveau bleu

#on  du dans les 3 dernieres instruction diviser sur 255 car les matrice 
#des images IMG1,IMG2,IMG3 sont du format entier, et imgRGB est flottante  
#dans ce cas la conversion du IMG1,IMG2,IMG3 est obligatoire ce qui est fait par division sur 255

plt.imsave("ImgCouleur1.bmp",imgRGB,cmap="gray") 
#sauvegarder l'image avec le titre "ImgCouleur.bmp" est utilis
#pour avoir l'image couleur resultante de cette cellule en format bmp 


plt.imshow(imgRGB)
plt.show()
```

### Etape 3 : afficher lhistogramme des images satellitaires en nuances de grise

Avec scikit-image nous utiliserons simplement la fonction histogram pour tracer ces graphes :

```python
def imageHist(image):
    _, axis = plt.subplots(ncols=2, figsize=(18, 3))#crer une figure et un ensemble de subplots
                                                    #figsize : la taille du figure
    if (image.ndim == 2): #Si on a 3 dimensions
        # Grascale Image
        axis[0].imshow(image, cmap=plt.get_cmap('gray'))
        axis[1].set_title('Histogram')#titre pour l'hstogramme
        axis[0].set_title('Grayscale Image')#titre pour l'image
        hist = exposure.histogram(image)#Return lhistogramme de limage.
        axis[1].plot(hist[0])
    else:
        # Color image
        axis[0].imshow(image, cmap='gray')
        axis[1].set_title('Histogram') #titre pour l'hstogramme
        axis[0].set_title('Colored Image') #titre pour l'image
        rgbcolors = ['red', 'green', 'blue'] #RGB
        for i, mycolor in enumerate(rgbcolors):
            axis[1].plot(exposure.histogram(image[...,i])[0], color=mycolor)  
            #Returnlhistogramme de limage.
            
    plt.legend(['Rouge','Vert','Bleu'])#fonction du legende (les attributs ont 
                                      # le meme ordre avec les 3 plots precedent)
    plt.xlabel('les intensit du couleur')
    plt.ylabel('les occurence des instensite')
```

Dans le code ci-dessus on regarde en premier lieu le nombre de dimensions de la matrice image. Si on a 3 dimensions cest que lon a une image avec couleur sinon on est en niveau de gris. Dans le cas o lon a une image avec couleur on doit empiler les canaux et donc nous aurons 3 courbes. comme suit :

###### Affichage lhistogramme aux niveaux de gris pour les trois images TM1, TM2 et TM3.
Une image en niveau de gris ne possde quune seule courbe (celle des nuances de gris):

```python
imageHist(TM1)
```

```python
imageHist(TM2)
```

```python
imageHist(TM3)
```

##### oui, possible de faire l'histogramme pour une image en couleur
une image en couleur il nous faut avoir les nuances sur les trois canaux (Rouge, Vert et Bleu). On a donc 3 courbes :

```python
IMGCLR=plt.imread("ImgCouleur1.bmp")
imageHist(IMGCLR)
```
